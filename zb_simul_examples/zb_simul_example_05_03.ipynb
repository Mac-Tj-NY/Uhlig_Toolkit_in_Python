```python
#   Uhlig's Exampl05.m  Model 3
#   This is  "Economy 3" from Farmer, R.E.A. 
#   and J.T. Guo, "Real Business Cycles and the 
#   Animal Spirits Hypothesis," Journal of Economic 
#  Theory 63, 42-72 (1994).

###  For instructional purposes Uhlig
###  set-up Exampl04.m  Model 3  incorrectly
###  and the model hangs when solved.
###  He shows the correct set-up in Exampl05  
###  by introducing an additional
###  state variable c(t) 

###  NOTE NOTE
### Py and MAT results match for Models 1 and ,
### but not model 3
###  differences in eigenvalues matter in Model 3


import numpy as np


#  These are the parameters for  
#  Farmer-Guo Model 3  

lamb = 0.58
b = 0.70
a = 0.23                
psi =1.0/1.05 
sigma_sigma = 0.001
psi_nu = 0
sigma_nu = 0.217

L_bar = 1.0/3
Z_bar = 1
delta = 0.025
rho = 0.99
eta = 1.0


alpha   = a/lamb
betta   = b/lamb  
R_bar = 1.0/rho
YK_bar = (R_bar + delta - 1)/a
K_bar = (YK_bar / (Z_bar*L_bar**betta) )**(1.0/(alpha-1))
I_bar = delta * K_bar
Y_bar = YK_bar * K_bar
C_bar = Y_bar - delta*K_bar
A_bar = C_bar**(-eta) * b * Y_bar/L_bar # Parameter in utility function


SUNSCALE = 2.0969
        

VARNAMES = ['capital',
                         'consumption',
                         'output',
                         'labor',
                         'interest',
                         'investment',
                         'technology',
                         'sunspot']

SteadyState = [[K_bar],
                         [C_bar],
                         [Y_bar],
                         [L_bar],
                         [R_bar],
                         [I_bar],
                         [1],
                         [SUNSCALE]]

AA = np.vstack(([0, -C_bar],
                          [- K_bar, 0],
                          [0, 0],
                          [0, -eta],
                          [0, 0]))

BB = np.vstack(([0, 0],
                          [(1-delta)*K_bar, 0],
                          [alpha, 0],
                          [0, 0],
                          [-a*YK_bar, 0]))

CC = np.vstack(([Y_bar, 0, 0, -I_bar],
                           [0, 0, 0, I_bar],
                           [-1, betta, 0, 0],
                           [1, -1, 0, 0],
                           [a*YK_bar, 0, -R_bar, 0]))

DD = np.vstack(([0,SUNSCALE],
                           [0,0],
                           [1,0],
                           [0,0],
                           [0,0]))

FF = [0, -eta]

GG = [0, eta]

HH = [0, 0]

JJ = [0, 0, 1, 0]

KK = [0, 0, 0, 0]

LL = [0,0]

MM = [0,0]

NN = np.vstack(([psi, 0],
                           [0, psi_nu]))

Sigma = np.vstack(([sigma_sigma**2, 0],
                                [0, sigma_nu**2]))

Sigma_EPS = Sigma


%run uhlig_solver_abreviated.ipynb



#print(' ')
#print('PP')
#print(PP.shape)
#print(PP)
#print(' ')
#print('RR')
#print(RR.shape)
#print(RR)
#print(' ')
#print('QQ')
#print(QQ.shape)
#print(QQ)
#print(' ')
#print('SS')
#print(SS.shape)
#print(SS)
#print(' ')


%run Uhlig_simul_rand.ipynb
```
